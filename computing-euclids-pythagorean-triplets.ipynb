### Formula
#### a = m^2 - n^2
#### b = 2mn
#### c = m^2 + n^2
#### m > n

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
number_range = np.arange(2,51)

a = []
b = []
c = []

for m in number_range:
    for n in range(number_range[0],m):
        a.append(m**2 - n**2)
        b.append(2*m*n)
        c.append(m**2 + n**2)
```

```python
fig,ax = plt.subplots(1,figsize=(10,7))

plt.scatter(a,b,c=c,alpha=.4,linewidths=.5)
plt.show()
```

### Verifying the elements in a,b, and c are valid tripplets should yield zeros

```python
path_check = np.array(a)**2 + np.array(b)**2 - np.array(c)**2

plt.plot(path_check,'o')
plt.show()
```
